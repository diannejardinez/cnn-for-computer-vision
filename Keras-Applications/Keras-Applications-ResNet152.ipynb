```python

# Dependencies
import matplotlib.pyplot as plt
%matplotlib inline

import os
import numpy as np
import tensorflow as tf

from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet import (ResNet152,
     preprocess_input, decode_predictions)

from keras import backend as K 
from keras.layers import Layer
```

```python
# Load the ResNet152 model
# https://keras.io/api/applications/resnet/#resnet152-function
model = ResNet152(
        include_top=True,
        weights="imagenet",
        input_tensor=None,
        input_shape=None,
        pooling=None,
        classes=1000
)
```

```python
# Image Size for ResNet152
image_size = (224, 224)
```

```python
# Prediction function using ResNet152 model
def predict(image_path):
    """Use ResNet152 to label image"""
    
    # Loading image for the input dimensions that ResNet152 was trained with
    img = image.load_img(image_path, target_size=image_size)
    
    # Preprocess image for model prediction (scaling and normalization for ResNet152)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    
    # Predictions
    predictions = model.predict(x)
    plt.imshow(img)
    print('Predicted:', decode_predictions(predictions, top=3)[0])
```

```python
image_path = os.path.join("Images", "image1.jpeg")
predict(image_path)
```
