```python

# Dependencies
import matplotlib.pyplot as plt
%matplotlib inline

import os
import numpy as np
import tensorflow as tf

from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.nasnet import (
    NASNetLarge, preprocess_input, decode_predictions)
```

```python
# Load the NASNetLarge model
# https://keras.io/api/applications/nasnet/#nasnetlarge-function
model = NASNetLarge(
    input_shape=None,
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    pooling=None,
    classes=1000,)
```

```python
# Default Image Size for NASNetLarge
image_size = (331, 331)
```

```python
# Prediction function using NASNetLarge model
def predict(image_path):
    """Use NASNetLarge to label image"""
    
    # Loading image for the input dimensions that NASNetLarge was trained with
    img = image.load_img(image_path, target_size=image_size)
    
    # Preprocess image for model prediction (scaling and normalization for NASNetLarge)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    
    # Predictions
    predictions = model.predict(x)
    plt.imshow(img)
    print('Predicted:', decode_predictions(predictions, top=5)[0])
```

```python
image_path = os.path.join("Images", "image1.jpeg")
predict(image_path)
```
